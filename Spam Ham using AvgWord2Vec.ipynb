```python
import pandas as pd

messages = pd.read_csv('SMSSpamCollection', sep='\t',
                           names=["label", "message"])

print(messages)
```

```python
import re 
import nltk
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
stop_words = set(stopwords.words("english")) 
```

```python
corpus = []
for i in range(len(messages)):
    sent = re.sub('[^a-zA-Z0-9]', ' ' ,messages['message'][i])
    sent = sent.lower()
    sent = sent.split()
    ## Stemming
    tokens = [ps.stem(word) for word in sent if not word in stop_words]
    sent = " ".join(tokens)
    corpus.append(sent)

corpus
```

```python
## creating BOW model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, binary=True, ngram_range=(2,2))
X = cv.fit_transform(corpus).toarray()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
messages = pd.read_csv('SMSSpamCollection', sep='\t',
                           names=["label", "message"])
```

```python
corpus = []
for i in range(len(messages)):
    review = re.sub('[^a-zA-Z0-9]', ' ', messages['message'][i])
    review = review.lower().split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in stop_words]
    review = " ".join(review)

    corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
len(corpus)
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
#     for token in sent_token:
    words.append(simple_preprocess(sent))
```

```python
import gensim
```

```python
model = gensim.models.Word2Vec(words, window=5, min_count=2)
```

```python
import numpy as np
## Avg word to vec
def avg_word2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key], axis = 0)
```

```python
# words
```

```python
from tqdm import tqdm
## Apply for entire sentences
X = []
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
```
