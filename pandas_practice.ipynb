#### Filter in pandas

```python

# axis=0 → rows, axis=1 → columns
# when selecting by labels (columns by name, pattern, etc.).
# Used mainly to filter rows/columns by labels (names, regex, lists).

df = titanic.filter(["Name", "Age", "Fare"])  ## Name, age, Fare
df = titanic.filter(like="me", axis=1)  ## column names like
df = titanic.filter(regex="^P",axis=1)
df
```

#### Query in pandas - 

```python
# when filtering rows by data values.
# Used to filter rows by conditions on column values.

df = titanic.query("Age > 25") ## all rows where age > 25
df = titanic.query("Sex == 'male'")  ## where sex is male
df = titanic.query("Age > 35 and Sex == 'male'")
df
```

```python
grouped = store_df.groupby("Category").agg({"Sales" : "sum", "Profit" : "mean", "Order ID":"count"}).rename(columns={"Order ID" : "Order_Count"})
print(grouped)
```

```python
## calculate profit margin - 
# def profit_margin(X):
#     return (X['Profit'].sum()/X['Sales'].sum())*100


# result = store_df.groupby("Category").agg({"Sales" : "sum", "Profit" : "sum"}).assign(Profit_margin=lambda x: (x['Profit']/x['Sales'])*100)

# print(result)


result = store_df.groupby("Category").agg(
    Total_sales=("Sales", "sum"), 
    Total_profit = ("Profit", "sum"))
 
print(result)
result = result.assign(profit_margin = (result['Total_profit'] / result['Total_sales'])*100)
    
result
```
