```python
## Always label based (whatever the label is) - slices are inclusive

df = titanic.loc[1:5] ## gets roews from 1st to 5th label
# df = titanic.loc[:, 4] ## does not work, loc in label based
df = titanic.loc[1:,] ## all rows from '1' label, all columns
df = titanic.loc[:5,] ## top 5 rows - upto label 5
# df
```

```python
df = titanic.loc[[1, 3, 4]]  # gets 1st, 3rd, 4th row
df = titanic.loc[1, "Name"]  # name of the passenger for label index 1 (two arguments only)
# df = titanic.loc[1, "Name", "Age"]  # FAILS - .loc[row, col] takes two arguments only:
df = titanic.loc[1, ["Name", "Age"]] # name, age of the passenger for label index 1
df = titanic.loc[:, ["Name", "Age"]] # all rows, these columns
df = titanic.loc[1:5, ["Name", "Age"]] # 1-5 rows(labels), these columns
df = titanic.loc[[1, 5], ["Name", "Age"]] # Name age for index 1 and 5(label)
df = titanic.loc[[1, 5], "Name":"Age"] # Name, Sex, age for index 1 and 5(label)
# df
```

```python
## boolean - 
df = titanic.loc[titanic["Age"] > 30, "Name"]  ## name with age > 30
df = titanic.loc[titanic["Age"] > 30, ["Name", "Age"]]  ## name, Age with age > 30
df
```

#### Filter in pandas

```python

# axis=0 → rows, axis=1 → columns
# when selecting by labels (columns by name, pattern, etc.).
# Used mainly to filter rows/columns by labels (names, regex, lists).

df = titanic.filter(["Name", "Age", "Fare"])  ## Name, age, Fare
df = titanic.filter(like="me", axis=1)  ## column names like
df = titanic.filter(regex="^P",axis=1)
df
```

#### Query in pandas - 

```python
# when filtering rows by data values.
# Used to filter rows by conditions on column values.

df = titanic.query("Age > 25") ## all rows where age > 25
df = titanic.query("Sex == 'male'")  ## where sex is male
df = titanic.query("Age > 35 and Sex == 'male'")
df
```
